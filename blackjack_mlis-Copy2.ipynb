```python
import random
import numpy as np
import pandas as pd
```

```python
suits=('Hearts','Diamonds','Spades','Clubs')
ranks=('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
values={'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}
playing=True
```

```python
class Card:
       
    def __init__(self,rank,suit):
        self.rank=rank
        self.suit=suit
        
    def __str__(self):
        return f'{self.rank} of {self.suit}'
```

```python
class Deck:
    
    def __init__(self):
        self.deck=[]
        
        for suit in suits:
            for rank in ranks:
            
                self.deck.append(Card(rank,suit))
        self.deck=n*self.deck
    
    def __iter__(self):
        return iter(self.deck)
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        self.deck.pop()
        return self.deck.pop()
```

```python
class Hand:
    def __init__(self):
        self.cards = []  
        self.value = 0   
        self.aces = 0    
    
    def add_card(self,card):
        self.cards.append(card)        
                                        
        self.value+=values[card.rank]
        if card.rank=='Ace':
            self.aces+=1
        
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()
```

```python
def hit_or_stand(deck, hand, action):
    global playing

    while True:
        if action == 'stand':
            playing = False
        elif action == 'hit':
            hit(deck,hand)
        break
```

```python
hits = np.zeros(len(states1))
stands = np.zeros(len(states1))
iterations = 100000
occurences = np.zeros(len(states1))

for i in range(iterations):

    n = 2
    the_deck = Deck()
    the_deck.shuffle()
    
    while True:    

        player_hand = Hand()

        player_hand.add_card(the_deck.deal())

        player_hand.add_card(the_deck.deal())

        player_hand.adjust_for_ace()
        
        reward_h = 0

        reward_s = 0
        
        while playing:

            if len(list(the_deck)) == 0:
                break 

            action = actions[random.randint(1,2)]
            previous_sum = player_hand.value
            
            hit_or_stand(the_deck, player_hand, action)

            if player_hand.value > 21:
                reward_h = -5
                break

        if player_hand.value <= 21:
            if action == 'hit':
                if player_hand.value >= previous_sum:
                    reward_h = 1
                    reward_s = -5
                    
                else:
                    reward_h = -1
            else:
                reward_s = 0.05
        
        for s in range(len(states1)):
            if previous_sum == states1[s][0]:
                if player_hand.aces == states1[s][1]:
                    hits[s] += reward_h
                    stands[s] += reward_s
                    occurences[s] += 1

        if len(list(the_deck)) > 4:
            playing = True
            continue
        else:
            break

for i in range(len(hits)):
    if occurences[i] != 0:
        hits[i] = hits[i]/occurences[i]
        stands[i] = stands[i]/occurences[i]
```
